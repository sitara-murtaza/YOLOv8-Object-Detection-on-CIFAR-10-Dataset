```python
!pip install ultralytics torchvision
from ultralytics import YOLO
from torchvision.datasets import CIFAR10
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
import cv2
import torch

# -------------------------------
# 1. Load YOLOv8 Nano Model
# -------------------------------
model = YOLO("yolov8n.pt")   # fastest YOLO

# -------------------------------
# 2. CIFAR-10 Class Names
# -------------------------------
cifar_classes = [
    'airplane', 'automobile', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

# -------------------------------
# 3. Image Transform
# -------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),   # upscale for YOLO
    transforms.ToTensor()
])

# -------------------------------
# 4. Load CIFAR-10 Dataset
# -------------------------------
dataset = CIFAR10(
    root="./data",
    train=False,
    download=True,
    transform=transform
)

# -------------------------------
# 5. Function: Convert Tensor ‚Üí OpenCV Image
# -------------------------------
def tensor_to_cv2(img_tensor):
    img = img_tensor.permute(1, 2, 0).numpy()   # CHW ‚Üí HWC
    img = (img * 255).astype(np.uint8)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    return img

# -------------------------------
# 6. Detect Images Category-Wise
# -------------------------------
images_per_class = 3   # show 3 images per category

for class_id, class_name in enumerate(cifar_classes):

    print(f"\nüîç Detecting category: {class_name.upper()}")

    count = 0

    for img_tensor, label in dataset:

        if label == class_id:
            count += 1

            # Convert image
            img = tensor_to_cv2(img_tensor)

            # Run YOLO detection
            results = model(
                img,
                conf=0.25,
                iou=0.45,
                max_det=1
            )

            # Plot result
            annotated_img = results[0].plot()

            # Display using matplotlib
            plt.figure(figsize=(4, 4))
            plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
            plt.title(f"CIFAR Class: {class_name}")
            plt.axis("off")
            plt.show()

            if count >= images_per_class:
                break
```
